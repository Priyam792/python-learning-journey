```python
import numpy as np
```

# Creating numpy arrays


```python
print(np.__version__)
arr = np.array([1, 2, 4, 5, 3])
arr1 = np.array(range(1, 11))
print(arr, arr1)
```

# Creating 2D arrays


```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([range(1, 10), range(11, 20), range(21, 30)])
print(a)
print("\n")
print(b)
```

# Array full of ZEROS

```python
a = np.zeros(5)
print(a)
b = np.zeros([5, 5])
print(b)
```

# Array full of ones


```python
a = np.ones(5)
print(a)
b = np.ones([5, 5])
print(b)
```

# Array full of ANY number


```python
a = np.full((2, 2), 3)
print(a)
```

# Identity matrix


```python
a = np.eye(7)
print(a)
```

# arange() {like range()}


```python
a = np.arange(1, 10).reshape(3, 3)  # <----- .reshape is used to shape the array
print(a)

```

# linspace


```python
a = np.linspace(1, 2, 4).reshape(2, 2)
print(a)
```

# **_CHECKING ARRAY PROPERTIES_**

```python
arr = np.array([[10, 20, 30], [40, 50, 60]])
```

### shape : used to identify the numbers of rows and column a array have

```python
arr.shape
```

### size : used to get total number of element in an array

```python
arr.size
```

### ndim : used to get dimensions of the array

```python
arr.ndim
```

### dtype : used to get the datatype of the elements inside the array

```python
arr.dtype
```

# **_Changing the data types_**

```python
arr = np.array([1, 2, 3], dtype=np.float32)
print(arr.dtype)
```

### astype  is used to change the datatype of an  existing array

```python
arr_int = arr.astype(np.int32)
print(arr_int)
```

# **_Reshaping and Flattening Arrays_**

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.shape)
```

```python
reshaped = arr.reshape((3, 2))
# Change shape
print(reshaped)
```

```python
reshaped.shape
```

### flatten is used to convert an array into 1D.

```python
flattened = arr.flatten()
print(flattened)
```

```python
print(flattened.shape)
flattened.ndim
```
