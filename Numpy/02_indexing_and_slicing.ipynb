# _**Slicing returns a view, not a copy! Changes affect the original array.**_

This might seem counterintuitive since Python lists create copies when sliced. But
in NumPy, slicing returns a view of the original array. Both the sliced array and the
original array share the same data in memory, so changes in the slice affect the
original array.

```python
a = flat[3:7]
```

```python
a
```

```python
a[0] = 67
```

```python
a
```

```python
flat
```

### here when i changed a but flat also change because slicing returns a view ,copy not a copy!

## Use .copy() if you need an independent copy.

```python
b = flat[3:9].copy()
```

```python
print(b)
flat
```

```python
b[0] = 943
```

```python
print(b)
flat
```

Not changed because of .copy()

# _**Fancy Indexing & Boolean Masking**_

# ------------------------------------------------------------------------

```python
# Fancy Indexing : select multiple element using index
ar = np.array([1,2,3,4,5,6,7,8,32,132,3,234,34])

print(ar[[1,3,6,8]])
```

```python
# Boolean Masking : filter data by condition
print(ar[ar>10])
```
