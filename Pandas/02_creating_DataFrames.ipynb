# _**Creating DataFrames**_
---
- Let’s look at different ways to create a Pandas DataFrame — the core data structure you’ll be using 90% of the time in data science.

```python
import pandas as pd , numpy as np
```

## From Python Lists

```python
data = [
    ["Priyam" , 65],
    ["Gary" , 98],
    ["Ash" , 87]
]
df = pd.DataFrame(data , columns=('Name','Marks') , index = range(1,4)) # Added custom index and column's name
df
```

## From Dictionary of lists
- Most common and readable format.
- Each key becomes a column, and each list is the column data.

```python
data = {
"Name": ["Alice", "Bob", "Charlie"],
"Age": [25, 30, 35]
}
df = pd.DataFrame(data)
df
```

## From NumPy Arrays
----

```python
arr = np.array([[1, 2], [3, 4]])
df = pd.DataFrame(arr)
df
```

# From CSV(Comma Seprated Values) Files
---

## 1. sep

- Specifies the delimiter used in the file.

```python
df = pd.read_csv("data.csv",sep=',')
df
```

## 2. header

- Specifies which row contains the column names.

```python
df = pd.read_csv("data.csv",header=1)
df
```

## 3. names

- Provide your own column names.

```python
df= pd.read_csv(
    "data.csv",
    # header=None , <--------- doesn't matter
    names=['Names','Movies','Time','Type','Collection','Rating']
)
df
```

## 4. index_col

- Choose which column becomes the index.

```python
df = pd.read_csv("data.csv", index_col="Film")
df
```

# 5. usecols

- Read only selected columns.

```python
# df = pd.read_csv("data.csv", usecols=["Actor","Film"])
df = pd.read_csv("data.csv", usecols=[0,1])  #<-------------------------------- Above one and this one works same

df
```

## _**Some more options**_
--
- nrows	Read    first n rows
- skiprows	    Skip initial rows
- dtype	Set     data types
- na_values	    Handle missing values
- parse_dates	Convert date columns

# _**EDA (Exploratory Data Analysis)**_
- It involves taking a deep look at the dataset to understand its structure, spot patterns, identify anomalies, and uncover relationships between variables.

## Why do EDA?

Suppose you receive a dataset with 10,000 rows.

Before analyzing it, you should know:

   -  How many rows and columns are there?

   -  Are there missing values?

   - Are there duplicate rows?

   - What are the data types?

   - Are there outliers?

   - What are the relationships between columns?

__That's what EDA helps you discover.__
