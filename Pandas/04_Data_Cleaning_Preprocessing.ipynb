Real-world data is messy. Pandas gives us powerful tools to clean and transform

### Handling Missing Values
---
Check for Missing Data

````python
import pandas as pd
````

````python
df = pd.read_csv("data-cleaning-sample.csv")
````

````python
df
````

````python
df.isnull() # Prints True for NaNs
````

````python
df.isnull().sum() # count missing per column
````

### Drop missing data

````python
df.dropna() # drop all the rows which have any missing values
````

````python
df.dropna(axis=1) # drop all the columns which have any missing values
````

### Fill Missing Data
- In pandas, fillna is used to fill unknown values. ffill and bfill are methods used to fill missing values (like NaN, None, or pd.NA) by propagating values forward or backward.

````python
df.fillna(0) # fills 0 in place of all missing values
````

````python
df["Age"].fillna(df["Age"].mean()).astype("int32") # Replace with mean also I changed it's datatype using .astype()
````

````python
df["Name"].ffill() # Forward fill > fill the previous value before the missing value
````

````python
df["Name"].bfill() # Backward fill > starts from last element and fill the next value to the missing value
````

### Detecting & Removing Duplicates

````python
df.duplicated() # Returns True for a duplicate **ROWS**
````

````python
df.drop_duplicates() # Remove duplicate **ROWS**
````

````python
df[["Name","Age"]].duplicated() # checks based on specific column
df.duplicated(subset=["Name", "Age"]) # Check based on specific column
````

## df.duplicated(subset=["Name", "Age"])
```

Pandas internally considers only the "Name" and "Age" columns while checking for duplicate rows.

It returns a Series of booleans.
```
---

## df[["Name", "Age"]].duplicated()

# Applying Functions
---

## .apply() → Apply any function to rows or columns

## .map() → Element-wise mapping for Series

## .replace() → Replace specific values


````python
df2["Age Group"] = df2["Age"].apply(lambda x : "adult" if x > 26 else "minor")
````

````python
df2
````

````python
gender_map = {"M" : "Male" , "F":"Female"}
df2["Gender"] = df2["Gender"].map(gender_map)
````

````python
df2
````

````python
replace_val = {"Mumbai" : "New Mumbai" , "New York" : "NYC"}
df2["City"] = df2["City"].replace(replace_val)
````
