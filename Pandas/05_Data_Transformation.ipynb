# Data Transformation
---
After datacleaning the next step is to reshape , reformat , and reorder it as needed for analysis. Pandas gives us plenty of flexible tools to do this. 

```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
df
```

### Sorting & Ranking

```python
df.sort_values("Year") # default in ascending
```

```python
df.sort_values("Year",ascending=False) # for descending
```

```python
df.sort_values(["Year","IMDb"]) # means if it finds year same it will move to IMDb . and ofc the order is ascending by default
```

### Reset index 
--- 
If you want the index to start from 0 and be sequential, you can reset it using reset_index()

```python
df2 = df.sort_values(["Year","IMDb"]).copy() # let's make new dataframe
```

```python
df2.reset_index()
```

### as you can see it makes the old index a column and added new index
### for removing it we use **drop=True**

```python
df2.reset_index(drop=True)
```

```python
df2
```

### But here you see it doesn't update the dataframe we are using .
### For updating it we use inplace= True

```python
df2.reset_index(drop=True,inplace=True)
df2
```

## Sort by Index

```python
df2.sort_index() # it will sort the rows according to the index . 
```

```python
df2["Rank"] = df2["IMDb"].rank() # But if there are same IMDb it'll give the average rank to both
df2
```

```python
# for not giving average we use method= 'dense' 
df2["Rank"] = df2["IMDb"].rank(method='dense')
df2
```

```python
df2.rename(columns={"Film": "Movie"}, inplace=True)
df2
```

```python
df.rename(index={0: "row1", 1: "row2"}, inplace=True) # it'll replace the index's name
df.head(3)
```

```python
df.columns = ["Name", "Movie", "Year" , "Type" , "Collection" , "Rating"]
df.head(1)
```

### Changing Column Order
- Just pass a new list of column names:

```python
df = df[["Movie","Name","Type","Year","Rating","Collection"]] # If you forget to mention any columns then it'll automatically drop
df.head(0)
```

- You can also move one column to the front:

```python
df.columns
```

```python
cols = ["Name"] + [col for col in df.columns if col != "Name"]
df= df[cols]
df.head(1)
```
